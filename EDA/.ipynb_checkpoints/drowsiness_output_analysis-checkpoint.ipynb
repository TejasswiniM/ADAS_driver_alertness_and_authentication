```python
import pandas as pd

# Replace 'output.csv' with your actual file name
df = pd.read_csv('output.csv')
```

```python
# Display basic statistics
#print(df.describe())

# Plot the trend of EAR over frames
import matplotlib.pyplot as plt

plt.plot(df['Frame no.'], df['EAR'])
plt.xlabel('Frame Number')
plt.ylabel('EAR')
plt.title('Eye Aspect Ratio Trend Over Frames')
plt.show()
```

```python
# Example: Identify frames where EAR is below a certain threshold
threshold_value = 0.25
drowsy_frames = df[df['EAR'] < threshold_value]
```

```python
# Example: Plot frames with low EAR values
drowsy_frames.plot(x='Frame no.', y='EAR', kind='scatter', color='red', label='Drowsy Frames')
plt.xlabel('Frame Number')
plt.ylabel('EAR')
plt.title('Frames with Drowsiness')
plt.show()
```

```python
# Assuming 'blink_threshold' is a suitable threshold for identifying blinks
blink_threshold = 0.2
frames_per_minute = 60

blink_frames = df[df['EAR'] < blink_threshold]

# Calculate blinking rate per minute
blinking_rate = len(blink_frames) / (len(df['Frame no.']) / frames_per_minute)
print(f'Blinking Rate: {blinking_rate:.2f} blinks per minute')
```

```python
# Assuming 'drowsiness_threshold' is a suitable threshold for identifying drowsiness
drowsiness_threshold = 0.3
drowsy_frames = df[df['EAR'] < drowsiness_threshold]

# Calculate duration of drowsy episodes
episode_durations = []
current_episode_start = None

for index, row in drowsy_frames.iterrows():
    if current_episode_start is None:
        current_episode_start = row['Frame no.']
    elif row['Frame no.'] - current_episode_start > 1:
        episode_durations.append(row['Frame no.'] - current_episode_start)
        current_episode_start = row['Frame no.']

average_duration = sum(episode_durations) / len(episode_durations) if episode_durations else 0
print(f'Average Duration of Drowsy Episodes: {average_duration:.2f} frames')
```
